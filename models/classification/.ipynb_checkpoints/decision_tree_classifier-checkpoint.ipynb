```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
```

```python
df = pd.read_csv("../../preprocessing/data_preprocessed.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
feature_names = df.columns.to_list()[:-1]
X = df[feature_names]
X
```

```python
X.describe()
```

```python
y = df['class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0, stratify = y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.classes_
```

```python
model.feature_importances_
```

```python
y_train_pred = model.predict(X_train)
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
y_test_pred = model.predict(X_test)
confusion_matrix(y_test, y_test_pred)
```

```python
plt.figure(figsize = (10, 10))
_ = plot_tree(model, feature_names=feature_names, class_names = model.classes_, filled = True)
```

```python

def decision_tree_report(model):

    report_str = (
        "Model Type: Decision Tree Classifier\n"
        f"Criterion: {model.criterion}\n"
        f"Splitter: {model.splitter}\n"
        f"Max Depth: {model.max_depth}\n"
        f"Min Samples Split: {model.min_samples_split}\n"
        f"Min Samples Leaf: {model.min_samples_leaf}\n"
        f"Min Impurity Decrease: {model.min_impurity_decrease}\n"
        f"Class Weight: {model.class_weight}\n"
        f"Random State: {model.random_state}"
    )

    return report_str
```

```python
print(decision_tree_report(model))
```

```python
params = {'criterion': ['gini', 'entropy'],
          'max_depth': range(2,10),
          'min_samples_leaf' : [2,3,4,5],
         }
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(), param_grid = params, scoring = 'accuracy', cv = 4, verbose = 4)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
estimator.best_score_
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=3, min_samples_leaf=2)
```

```python
model.fit(X_train, y_train)
```

```python
attributes_report = decision_tree_report(model)
print(attributes_report)
```

```python
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)
```

```python
def random_forest_report(model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))
```

```python
random_forest_report(model, X_test, y_test)
```
