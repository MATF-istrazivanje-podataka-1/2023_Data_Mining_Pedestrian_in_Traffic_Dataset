```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("../../preprocessing/data_preprocessed.csv")
```

```python
df.head()
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)
```

```python
pca = PCA(n_components = 3)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
```

```python
pca_names = [f'pca_{i}' for i in range(3)]
X_pca = pd.DataFrame(X_pca, columns = pca_names)
X_pca
```

```python
model = AgglomerativeClustering(n_clusters=3, linkage='single', compute_distances=True)
```

```python
model.fit(X_pca)
```

```python
model.labels_
```

```python
np.unique(model.labels_)
```

```python
model.children_.shape
```

```python
model.distances_.reshape(-1,1).shape
```

```python

n_clusters_list = [2, 3, 4, 5, 6]  
linkages = ['single', 'complete', 'average']
fig = plt.figure(figsize=(15, 15))

for n_clusters in n_clusters_list:
    for idx, linkage_type in enumerate(linkages):
        model = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage_type)
        model.fit(X_pca)

        ax = fig.add_subplot(len(n_clusters_list), len(linkages), (n_clusters_list.index(n_clusters) * len(linkages)) + idx + 1, projection='3d')

        for cluster_label in range(n_clusters):
            cluster = X_pca[model.labels_ == cluster_label]
            ax.scatter(cluster['pca_0'], cluster['pca_1'], cluster['pca_2'], label=f'Cluster {cluster_label}')

        ax.set_title(f'n_clusters: {n_clusters}, Linkage: {linkage_type}\nSilhouette Score: {silhouette_score(X_pca, model.labels_)}', y=1.05)  # Adjust the 'y' coordinate to move the title
        ax.legend()

plt.tight_layout()
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

def plot_dendrograms(X_pca):
    for column_name in X_pca.columns:
        X_pca_copy = X_pca.copy()  # Make a copy of the original DataFrame
        # Create linkage matrix
        Z = linkage(X_pca_copy[[column_name]], method='average')
        
        # Plot the dendrogram
        plt.figure(figsize=(10, 5))
        plt.title(f'Dendrogram for {column_name}')
        plt.xlabel('Data Points')
        plt.ylabel('Distance')
        dendrogram(Z, leaf_rotation=90, no_labels=True)
        plt.show()
```

```python
plot_dendrograms(X_pca)
```
